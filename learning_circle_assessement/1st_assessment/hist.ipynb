```python
import matplotlib.pyplot as plt
import numpy as np

# Data
intervals = ["5.0 - 5.2", "5.2 - 5.4", "5.4 - 5.6", "5.6 - 5.8", "5.8 - 6.0"]
frequencies = [14, 52, 51, 27, 11]

# Calculate midpoints of intervals
midpoints = [float(interval.split()[0]) + 0.1 for interval in intervals]

# Plotting the histogram bars
plt.bar(intervals, frequencies, color='blue', alpha=0.7, label='Histogram')

# Plotting the continuous curve through midpoints
curve_fit = np.polyfit(midpoints, frequencies, 2)
curve_x = np.linspace(min(midpoints), max(midpoints), 100)
curve_y = np.polyval(curve_fit, curve_x)

# Plotting the continuous curve on top of the histogram
plt.plot(curve_x, curve_y, color='red', label='Continuous Curve')

# Adding labels and title
plt.xlabel('Intervals')
plt.ylabel('Frequency')
plt.title('Histogram with Continuous Curve')

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```

```python
# Data points
midpoints = [5.1, 5.3, 5.5, 5.7, 5.9]
frequencies = [14, 52, 51, 27, 11]

# Assuming uniform interval width
interval_width = 0.2

# Calculate the area using the trapezoidal rule
area = (interval_width / 2) * (frequencies[0] + 2 * sum(frequencies[1:-1]) + frequencies[-1])

print("Estimated Area under the Curve:", area)
```

```python
# Data points
midpoints = [5.1, 5.3, 5.5, 5.7, 5.9]
frequencies = [14, 52, 51, 27, 11]

# Calculate the mean
mean_value = sum(midpoint * frequency for midpoint, frequency in zip(midpoints, frequencies)) / sum(frequencies)

# Calculate the absolute deviations
absolute_deviations = [abs(midpoint - mean_value) for midpoint in midpoints]

# Calculate the mean absolute deviation
mean_absolute_deviation = sum(absolute_deviations) / len(midpoints)

print("Mean Absolute Deviation:", mean_absolute_deviation)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

# Given data
Mx_over_x = np.array([9.30, 9.85, 8.46, 8.93, 8.08, 7.67, 10.40, 5.00])
c = np.array([0.460, 0.390, 0.209, 0.150, 0.08, 0.046, 0.026, 0.005])

# Define the Langmuir isotherm equation
def langmuir_isotherm(c, K, x_m_max):
    return (K * c) / (1 + K * c / x_m_max)

# Fit the Langmuir isotherm to the data
params, covariance = curve_fit(langmuir_isotherm, c, Mx_over_x)

# Extract the fitted parameters
K, x_m_max = params

# Generate points for the fitted curve
c_fit = np.linspace(min(c), max(c), 100)
Mx_over_x_fit = langmuir_isotherm(c_fit, K, x_m_max)

# Plotting
plt.scatter(c, Mx_over_x, label='Data points')
plt.plot(c_fit, Mx_over_x_fit, label='Langmuir Fit', color='red')
plt.title('Langmuir Isotherm Fit')
plt.xlabel('c')
plt.ylabel('Mx/x')
plt.legend()
plt.show()
```

```python
import pandas as pd

# Create two DataFrames
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [2, 3, 4], 'Age': [25, 30, 22]})

# Inner join on 'ID'
result_inner = pd.merge(df1, df2, on='ID', how='inner')
print(result_inner)
```

```python
# Left join on 'ID'
result_left = pd.merge(df1, df2, on='ID', how='left')
print(result_left)
```

```python
# Right join on 'ID'
result_right = pd.merge(df1, df2, on='ID', how='right')
print(result_right)
```
